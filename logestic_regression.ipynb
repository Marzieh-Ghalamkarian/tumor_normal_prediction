```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from sklearn import preprocessing
import scipy.optimize as opt
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv((r'C:\Users\Ghalamkarian\PycharmProjects\pythonProject3\data_breastcancer.csv'))
```

```python
ya=data['diagnosis']
le=LabelEncoder()
encoderdata4=le.fit_transform(ya)
data['diagnosis']=encoderdata4
```

```python
data.tail()
```

```python
X=data[['radius_mean','texture_mean','perimeter_mean','smoothness_mean','compactness_mean']]#.values
x=np.asarray(X)
x[0:5]
```

```python
Y=data['diagnosis']
y=np.asarray(Y)
y[0:5]
Y.value_counts()
```

```python
#normalize
scalearn=preprocessing.StandardScaler()
x=scalearn.fit(x).transform(x)
x[0:5]
```

```python
#train & test 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
x_test
x_train
```

```python
data.size
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
lr=LogisticRegression(C=0.01).fit(x_train,y_train)
lr
```

```python
yhat=lr.predict(x_test)
yhat.reshape
print(yhat[0:10])
print(y_test[0:10])
```

```python
yhat_proba=lr.predict_proba(x_test)
yhat_proba[0:10]
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat, pos_label=1)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat, pos_label=0)
```
